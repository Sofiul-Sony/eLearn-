```python
import numpy as np
DataYes = np.loadtxt('FeatureYes.txt',dtype=float)
DataNo = np.loadtxt('FeaturesNo.txt',dtype=float)
```

```python
from sklearn.utils import shuffle
DataX = np.array(DataYes.tolist()  + DataNo.tolist() )
DataY = np.array(np.ones(len(DataYes)).tolist() + np.zeros(len(DataNo)).tolist())
DataX , DataY = shuffle(DataX , DataY)
```

```python
columns = ['f' + str(i) for i in range(512)]
```

```python
columns.append('label')
```

```python
import pandas as pd
df = pd.DataFrame(columns = columns)

    
    
```

```python
df
```

```python
i = 0
for item in DataYes:
    #print(item.tolist())
    j = 0
    for index in  item:
        feature = 'f'+str(j)
        df.loc[i, feature] = index
        j = j + 1
    df.loc[i, 'label'] = 1
    i = i+1
```
